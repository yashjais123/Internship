# A) Rank
# B) Name
# C) Artist
# D) Upload date

```python
driver = webdriver.Edge(r"C:/Users/yjjai/Downloads/edgedriver_win64/msedgedriver.exe")
```

```python
driver.maximize_window()
```

```python
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
```

```python
Rank = []
Name = []
Artist = []
Upload_Date = []
Views = []



try:
    for i in driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]//tbody//tr//td[1]')[:30]:
        Rank.append(i.text)
except NoSuchElementException:
    Rank.append('-')

    
try:
    for i in driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]//tbody//tr//td[2]')[:30]:
        Name.append(i.text)
except NoSuchElementException:
    Name.append('-')

    
try:
    for i in driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]//tbody//tr//td[3]')[:30]:
        Artist.append(i.text)
except NoSuchElementException:
    Artist.append('-')

    
try:
    for i in driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]//tbody//tr//td[5]')[:30]:
        Upload_Date.append(i.text)
except NoSuchElementException:
    Upload_Date.append('-')

    
try:
    for i in driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]//tbody//tr//td[4]')[:30]:
        Views.append(i.text)
except NoSuchElementException:
    Views.append('-')
```

```python
import pandas as pd
most_viewed = pd.DataFrame()
most_viewed['Rank'] = Rank
most_viewed['Name'] = Name
most_viewed['Artist'] = Artist
most_viewed['Date'] = Upload_Date
most_viewed['Views'] = Views
most_viewed 
```

```python
Dataset_name = []
Data_type = []
Task = []
Attribute_type = []
No_of_instances = []
No_of_attribute = []
Year = []


N_name = driver.find_elements(By.XPATH,'//p[@class="normal"]//b//a')
for name in N_name:
    Dataset_name.append(name.text)
    
#scraping the data of data
try:
    op_data = driver.find_elements(By.XPATH,'//table[2]//tbody[1]//tr[1]//td[3]//p[@class="normal"]')
    for os in op_data:
        Data_type.append(os.text)
except NoSuchElementException:
    Data_type.append('-')


#scraping data of display of the Task
try:
    task = driver.find_elements(By.XPATH,'//table[2]//tbody[1]//tr[1]//td[4]//p[@class="normal"]')
    for disp in task:
        Task.append(disp.text)
except NoSuchElementException:
    Task.append("-")


# scraping data of Attribute type
try:
    att = driver.find_elements(By.XPATH,'//table[2]//tbody[1]//tr[1]//td[4]//p[@class="normal"]')
    for pro in att:
        Attribute_type.append(pro.text)
except NoSuchElementException:
     Attribute_type.append('-')


# scraping the data of No_of_instances
try:
    inst = driver.find_elements(By.XPATH,'//table[2]//tbody[1]//tr[1]//td[5]//p[@class="normal"]')
    for memo in inst:
        No_of_instances.append(memo.text)
except NoSuchElementException:
    No_of_instances.append('-')
    
    
    
# scraping the data of No_of_attribute
try:
    attri = driver.find_elements(By.XPATH,'//table[2]//tbody[1]//tr[1]//td[6]//p[@class="normal"]')
    for memo in attri:
        No_of_attribute.append(memo.text)
except NoSuchElementException:
    No_of_attribute.append('-')   
    
    
# scraping the data of Year
try:
    year = driver.find_elements(By.XPATH,'//table[2]//tbody[1]//tr[1]//td[7]//p[@class="normal"]')
    for mem in year:
        Year.append(mem.text)
except NoSuchElementException:
    Year.append('-')   
```

```python
import pandas as pd
UCI = pd.DataFrame()
UCI['Name'] = Dataset_name
UCI['Data type'] = Data_type
UCI['Task'] = Task
UCI['Attribute_type'] = Attribute_type
UCI['No_of_instances'] = No_of_instances
UCI['No_of_attribute'] = No_of_attribute
UCI['Year'] = Year
UCI
```
