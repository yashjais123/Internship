```python
import selenium
from selenium import webdriver 
import pandas as pd
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings('ignore')
import time

from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
driver = webdriver.Edge(r"C:/Users/yjjai/Downloads/edgedriver_win64/msedgedriver.exe")
driver.maximize_window()
```

```python
url = "https://flight.yatra.com/air-search-ui/dom2/trigger?type=O&viewName=normal&flexi=0&noOfSegments=1&origin=DEL&originCountry=IN&destination=BOM&destinationCountry=IN&flight_depart_date=04%2F01%2F2023&ADT=1&CHD=0&INF=0&class=Economy&source=fresco-home&unqvaldesktop=983293448512"
driver.get(url)
```

```python
Airline_name = []
Depart_time = []
Arival_time = []
Duration = []
Stop_time = []
Price = []
From = []
To = []

#scraping the data of Airline_name
try:
    air_name = driver.find_elements(By.XPATH,'//span[@class="i-b text ellipsis"]')
    for os in air_name:
        Airline_name.append(os.text)
except NoSuchElementException:
    Airline_name.append('-')

    
#scraping the data of Depart_time
try:
    air_dec = driver.find_elements(By.XPATH,'//div[@class="i-b pr"]')
    for os in air_dec:
        Depart_time.append(os.text)
except NoSuchElementException:
    Depart_time.append('-')


#scraping data of display of the Arival_time
try:
    air_ari = driver.find_elements(By.XPATH,'//p[@class="bold fs-15 mb-2 pr time"]')
    for disp in air_ari:
        Arival_time.append(disp.text)
except NoSuchElementException:
    Arival_time.append("-")


# scraping data of Duration
try:
    dur = driver.find_elements(By.XPATH,'//p[@class="fs-12 bold du mb-2"]')
    for pro in dur:
        Duration.append(pro.text)
except NoSuchElementException:
    Duration.append('-')
    
    
# scraping data of Stop_time
try:
    stop = driver.find_elements(By.XPATH,'//div[@class=" font-lightgrey fs-10 tipsy i-b fs-10"]')
    for pro in stop:
        Stop_time.append(pro.text)
except NoSuchElementException:
    Stop_time.append('-')
    
    
#scraping the data of Price
try:
    op_ser = driver.find_elements(By.XPATH,'//div[@class="i-b tipsy fare-summary-tooltip fs-18"]')
    for os in op_ser:
        Price.append(os.text)
except NoSuchElementException:
    Price.append('-')
    
    
# scraping  Flight Name
try:
    name = driver.find_elements(By.XPATH,"//p[1][@class='fs-10 font-lightgrey no-wrap city ellipsis']")
    for i in name:
        From.append(i.text)
except NoSuchElementException:
    From.append('-')
    
    
# scraping  Flight Name
try:
    name = driver.find_elements(By.XPATH,"//p[2][@class='fs-10 font-lightgrey no-wrap city ellipsis']")
    for i in name:
        To.append(i.text)
except NoSuchElementException:
    To.append('-')   
```

```python
import pandas as pd
flight_1 = pd.DataFrame()
flight_1['Airline name'] = Airline_name
flight_1['From'] = From
flight_1['To'] = To
flight_1['Arival time'] = Arival_time
flight_1['Depart time'] = Depart_time
flight_1['Stop time'] = Stop_time
flight_1['Duration'] = Duration
flight_1['Price'] = Price
flight_1 
```

```python
flight_1.to_csv('flight_1')
```

```python
Airline_name = []
Depart_time = []
Arival_time = []
Duration = []
Stop_time = []
Price = []
From = []
To = []

#scraping the data of Airline_name
try:
    air_name = driver.find_elements(By.XPATH,'//span[@class="i-b text ellipsis"]')
    for os in air_name:
        Airline_name.append(os.text)
except NoSuchElementException:
    Airline_name.append('-')

    
#scraping the data of Depart_time
try:
    air_dec = driver.find_elements(By.XPATH,'//div[@class="i-b pr"]')
    for os in air_dec:
        Depart_time.append(os.text)
except NoSuchElementException:
    Depart_time.append('-')


#scraping data of display of the Arival_time
try:
    air_ari = driver.find_elements(By.XPATH,'//p[@class="bold fs-15 mb-2 pr time"]')
    for disp in air_ari:
        Arival_time.append(disp.text)
except NoSuchElementException:
    Arival_time.append("-")


# scraping data of Duration
try:
    dur = driver.find_elements(By.XPATH,'//p[@class="fs-12 bold du mb-2"]')
    for pro in dur:
        Duration.append(pro.text)
except NoSuchElementException:
    Duration.append('-')
    
    
# scraping data of Stop_time
try:
    stop = driver.find_elements(By.XPATH,'//div[@class=" font-lightgrey fs-10 tipsy i-b fs-10"]')
    for pro in stop:
        Stop_time.append(pro.text)
except NoSuchElementException:
    Stop_time.append('-')
    
    
#scraping the data of Price
try:
    op_ser = driver.find_elements(By.XPATH,'//div[@class="i-b tipsy fare-summary-tooltip fs-18"]')
    for os in op_ser:
        Price.append(os.text)
except NoSuchElementException:
    Price.append('-')
    
    
# scraping  Flight Name
try:
    name = driver.find_elements(By.XPATH,"//p[1][@class='fs-10 font-lightgrey no-wrap city ellipsis']")
    for i in name:
        From.append(i.text)
except NoSuchElementException:
    From.append('-')
    
    
# scraping  Flight Name
try:
    name = driver.find_elements(By.XPATH,"//p[2][@class='fs-10 font-lightgrey no-wrap city ellipsis']")
    for i in name:
        To.append(i.text)
except NoSuchElementException:
    To.append('-')   
```

```python
import pandas as pd
flight_9 = pd.DataFrame()
flight_9['Airline name'] = Airline_name
flight_9['From'] = From
flight_9['To'] = To
flight_9['Arival time'] = Arival_time
flight_9['Depart time'] = Depart_time
flight_9['Stop time'] = Stop_time
flight_9['Duration'] = Duration
flight_9['Price'] = Price
flight_9 
```

```python
flight_9.to_csv('flight_9')
```

```python
Airline_name = []
Depart_time = []
Arival_time = []
Duration = []
Stop_time = []
Price = []
From = []
To = []

#scraping the data of Airline_name
try:
    air_name = driver.find_elements(By.XPATH,'//span[@class="i-b text ellipsis"]')
    for os in air_name:
        Airline_name.append(os.text)
except NoSuchElementException:
    Airline_name.append('-')

    
#scraping the data of Depart_time
try:
    air_dec = driver.find_elements(By.XPATH,'//div[@class="i-b pr"]')
    for os in air_dec:
        Depart_time.append(os.text)
except NoSuchElementException:
    Depart_time.append('-')


#scraping data of display of the Arival_time
try:
    air_ari = driver.find_elements(By.XPATH,'//p[@class="bold fs-15 mb-2 pr time"]')
    for disp in air_ari:
        Arival_time.append(disp.text)
except NoSuchElementException:
    Arival_time.append("-")


# scraping data of Duration
try:
    dur = driver.find_elements(By.XPATH,'//p[@class="fs-12 bold du mb-2"]')
    for pro in dur:
        Duration.append(pro.text)
except NoSuchElementException:
    Duration.append('-')
    
    
# scraping data of Stop_time
try:
    stop = driver.find_elements(By.XPATH,'//div[@class=" font-lightgrey fs-10 tipsy i-b fs-10"]')
    for pro in stop:
        Stop_time.append(pro.text)
except NoSuchElementException:
    Stop_time.append('-')
    
    
#scraping the data of Price
try:
    op_ser = driver.find_elements(By.XPATH,'//div[@class="i-b tipsy fare-summary-tooltip fs-18"]')
    for os in op_ser:
        Price.append(os.text)
except NoSuchElementException:
    Price.append('-')
    
    
# scraping  Flight Name
try:
    name = driver.find_elements(By.XPATH,"//p[1][@class='fs-10 font-lightgrey no-wrap city ellipsis']")
    for i in name:
        From.append(i.text)
except NoSuchElementException:
    From.append('-')
    
    
# scraping  Flight Name
try:
    name = driver.find_elements(By.XPATH,"//p[2][@class='fs-10 font-lightgrey no-wrap city ellipsis']")
    for i in name:
        To.append(i.text)
except NoSuchElementException:
    To.append('-')   
```

```python
import pandas as pd
flight_17 = pd.DataFrame()
flight_17['Airline name'] = Airline_name
flight_17['From'] = From
flight_17['To'] = To
flight_17['Arival time'] = Arival_time
flight_17['Depart time'] = Depart_time
flight_17['Stop time'] = Stop_time
flight_17['Duration'] = Duration
flight_17['Price'] = Price
flight_17 
```

# convert object columns are int64 columns

```python
df['Price'] = df['Price'].str.replace(',','').astype(int)
```

```python
df.head()
```

```python
# Depart time is when a plane leaves the gate. 

# Extracting Hours
df["Depart hour"] = pd.to_datetime(df["Depart_time"]).dt.hour

# Extracting Minutes
df["Depart min"] = pd.to_datetime(df["Depart_time"]).dt.minute

# Now we can drop Dep_Time as it is of no use
df.drop(["Depart_time"], axis = 1, inplace = True)
```

```python
df.head()
```

```python
# Assigning and converting Duration column into list
duration = list(df["Duration"])

for i in range(len(duration)):
    if len(duration[i].split()) != 2:    # Check if duration contains only hour or mins
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m"   # Adds 0 minute
        else:
            duration[i] = "0h " + duration[i]           # Adds 0 hour

duration_hours = []
duration_mins = []
for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = "h")[0]))    # Extract hours from duration
    duration_mins.append(int(duration[i].split(sep = "m")[0].split()[-1]))   # Extracts only minutes from duration
```

```python
# Adding duration_hours and duration_mins list to train_data dataframe

df["Duration_hours"] = duration_hours
df["Duration_mins"] = duration_mins
df.drop(["Duration"], axis = 1, inplace = True)
```
